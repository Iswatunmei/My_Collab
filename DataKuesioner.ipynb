```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
file_path = "/content/KUESIONER_EVALUASI.csv"
df = pd.read_csv(file_path, encoding="latin1")

# ---------------------------------
# Analisis Korelasi
# ---------------------------------
print("\nAnalisis Korelasi")
num_cols = df.select_dtypes(include=['number'])

# Hitung matriks korelasi
correlation_matrix = num_cols.corr()

# Visualisasi heatmap korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Heatmap Korelasi")
plt.show()

print("Analisis korelasi selesai.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
file_path = "/content/KUESIONER_EVALUASI.csv"
df = pd.read_csv(file_path, encoding="latin1")

# ---------------------------------
# Analisis Multivarian dengan Pairplot
# ---------------------------------
print("\nAnalisis Multivarian")

# Pilih beberapa variabel numerik untuk pairplot
plt.figure(figsize=(10, 8))
num_vars = ["Aplikasi Access by KAI mudah digunakan.", "Saya merasa aman dalam melakukan transaksi melalui aplikasi ini.", "Informasi yang disajikan dalam aplikasi akurat dan relevan."]
for var in num_vars:
    if var not in df.columns:
        print(f"Variabel {var} tidak ditemukan dalam dataset.")
        break
else:
    sns.pairplot(df[num_vars])
    plt.show()

print("Analisis multivarian selesai.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
file_path = "/content/KUESIONER_EVALUASI.csv"
df = pd.read_csv(file_path, encoding="latin1")

# ---------------------------------
# Analisis Numerik vs Kategorikal
# ---------------------------------
print("\nAnalisis Numerik vs Kategorikal")

num_var = "Aplikasi Access by KAI mudah digunakan."
cat_var = "Fitur apa yang paling sering Anda gunakan di aplikasi Access by KAI?"

if num_var in df.columns and cat_var in df.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[cat_var], y=df[num_var])
    plt.title(f"Boxplot {num_var} berdasarkan {cat_var}")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Salah satu variabel tidak ditemukan dalam dataset.")

print("Analisis numerik vs kategorikal selesai.")
```
